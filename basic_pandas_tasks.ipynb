# Basic Pandas Tasks

This notebook demonstrates basic operations using the Pandas library in Python.

```python
# Import necessary libraries
import pandas as pd
```

## Creating a DataFrame

Below is an example of creating a DataFrame from a dictionary.

```python
# Create a simple DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [24, 27, 22, 32],
    'City': ['New York', 'Los Angeles', 'Chicago', 'Houston']
}
df = pd.DataFrame(data)
df
```

## Basic DataFrame Operations

Demonstrating how to select columns and filter rows in the DataFrame.

```python
# Select the 'Name' column
names = df['Name']
names
```

```python
# Filter rows where age is greater than 25
older_than_25 = df[df['Age'] > 25]
older_than_25
```

## Descriptive Statistics

Get summary statistics of the DataFrame.

```python
# Get descriptive statistics
df.describe()
```

## Data Manipulation

Learn how to add and remove columns from the DataFrame.

```python
# Adding a new column
df['Profession'] = ['Engineer', 'Doctor', 'Artist', 'Lawyer']
df
```

```python
# Removing the 'City' column
df = df.drop('City', axis=1)
df
```

## Sorting Data

Sort the DataFrame by age.

```python
# Sort by age
df_sorted = df.sort_values(by='Age')
df_sorted
```

## Data Visualization

Visualizing data using Pandas' built-in plotting functionality.

```python
# Simple plot example
df_sorted.plot(x='Name', y='Age', kind='bar')
```
