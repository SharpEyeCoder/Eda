# Data Visualization Example
This Jupyter Notebook demonstrates various data visualization techniques using Python libraries such as `pandas`, `matplotlib`, and `seaborn`.

```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
sns.set(style='whitegrid') # Set style
```

## Load Sample Data
We will use the `iris` dataset, which is commonly used for visualization examples.

```python
# Load iris dataset
iris = sns.load_dataset('iris')
iris.head()
```

## Scatter Plot
A scatter plot shows the relationship between two variables.

```python
# Create a scatter plot
sns.scatterplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Scatter Plot of Sepal Length vs Sepal Width')
plt.show()
```

## Histogram
Histograms show the distribution of a variable.

```python
# Create a histogram
sns.histplot(data=iris, x='petal_length', bins=20, hue='species', kde=True)
plt.title('Histogram of Petal Length')
plt.show()
```

## Box Plot
Box plots are used to show the distribution of data and identify any outliers.

```python
# Create a box plot
sns.boxplot(data=iris, x='species', y='petal_width')
plt.title('Box Plot of Petal Width by Species')
plt.show()
```

## Line Plot
Line plots are great for visualizing trends over time or ordered data.

```python
# Create a line plot
sns.lineplot(data=iris, x='sepal_length', y='sepal_width', hue='species')
plt.title('Line Plot of Sepal Length vs Sepal Width')
plt.show()
```
